### 1. Identify the problem and how to address it using machine learning

ACME is committed to sustainability and has made efforts to reduce energy consumption in its ships. Energy efficient equipment was installed in 9 test rooms that have been operating for ~5 months from January 11 2016 to May 27 2016, collecting 10 minutes readings of total energy consumption (of all 9 rooms combined), room temperature, room humidity, and their outside counterparts along with outside windspeed, visibility, and pressure.

The request to the data science team is an algorithm that can accurately predict the energy consumption of rooms so that environmental and financial savings can be estimated for complete ships, or even fleets, before making the expensive renovations.

The task is to leverage regression machine learning algorithms to predicts the target variable, total energy consumption (TotalConsmp), with the predictor variable (inside temperature, inside humidity, outside temperature, outside humidity, outside pressure, visibility and windspeed. 


multivariate multi-step time series forecasting model

Being able to predict a month ahead with 5-month prior, or 1 week ahead with 6-week prior

The performance metric for this problem will be the RMSE 

### 2. Stats on energy consumption at different times (Monthly, Weekly, Day vs Night, Weekends vs Weekdays) 

````python
# summary statistics about the data
data.describe()
````

Getting a feel for the data by going through the max, min, median, and quantiles manually. However manual checks are not efficient, let's make some plots. Thorough Data Cleanup will be done in the Pre-Processing notebook.

````python
# Checking energy consumption over time, great variance.
# Hypothesis: Variance could depend on on-sea vs docked, or occupied vs unoccupied, or day vs night.

data.TotalConsmp.plot(y='TotalConsmp', figsize=(18, 6), )
plt.ylabel('Consumption kWh')
````

````python
# checking for repeated 10min intervals in the index. All good.
data.index[data.index.duplicated()].unique()
````

````python
# Summary stats of Total Consumption
data['TotalConsmp'].describe()
````

````python
# frequency was downsampled from 10 min, to 1 month, summing each month's consumption values. 
# data starts on January 11 2016 and ends on May 27 2016

monthly_consumption = pd.DataFrame(data['TotalConsmp'].resample('M').sum())

monthly_consumption.divide(1000).plot(kind='bar', figsize=(8, 6))
plt.title('Monthly Energy Consumption â€” All Rooms', fontsize=16)
plt.ylabel('Consumption in KWh', fontsize=14)

monthly_consumption 
````

````python
weekly_consumption = pd.DataFrame(data['TotalConsmp'].resample('W').sum())

weekly_consumption.divide(1000).plot(kind='bar', figsize=(8, 6))
plt.title('Weekly Energy Consumption â€” All Rooms', fontsize=16)
plt.ylabel('Consumption in KWh', fontsize=14)

weekly_consumption
````

````python
# defined 'day' to be from 6:00am to 6:00pm, this can be revised later
data['isDay'] = False
data.loc[data.between_time('6:00', '17:59').index, 'isDay'] = True
````

````python
data[['TotalConsmp', 'isDay']].groupby('isDay').sum().divide(1000).plot(kind='bar')
plt.title('Day vs Night Consumption â€” All Rooms', fontsize=16)
plt.ylabel('Consumption in KWh', fontsize=14)
pd.DataFrame(data[['TotalConsmp', 'isDay']].groupby('isDay').sum())
````

````python
# weekday returns the day of the week 0 = Monday, 6 = Sunday

data['isWeekDay'] = False
data.loc[data.index.weekday < 5, 'isWeekDay'] = True
````

````python
data[['TotalConsmp', 'isWeekDay']].groupby('isWeekDay').sum().divide(1000).plot(kind='bar')
plt.title('Weekday vs Weekend Consumption â€” All Rooms', fontsize=16)
plt.ylabel('Consumption in KWh', fontsize=14)

pd.DataFrame(data[['TotalConsmp', 'isWeekDay']].groupby('isWeekDay').sum())

# expected as there are 5 weekdays and 2 weekends
````

### 3. Effect of temp in energy consumption.

````python
# making utility variables for room temp and room humidity
r_temps = [t for t in data.columns if 'R' in t]
r_humids = [h for h in data.columns if 'H_' in h and len(h) < 4]
````

````python
# data is resampled from 10min intervals to hourly to improve graph clarity and load time.

ax = data[r_temps + ['TempOutSide']].resample('H').mean().plot(figsize=(18,6), legend=False)
ax2 = ax.twinx()
data.TotalConsmp.resample('H').sum().plot(ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.title('Total Consumption vs Temperature Outside and Room Temperatures')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data[['TotalConsmp', 'TempOutSide'] + r_temps].corr())
````

This correlation heatmap carries a lot of interesting information. To start, all the Room temeperatures have high covariance (they vary together i.e. are correlated), thus they do not provide new information separately and will likely be combined via Feature Engineering in the Pre-Processing notebook. 

Room 6 temperature is heavily correlated to the outside temperature. Maybe this sensor was placed outside the room or the room was left open and/or unoccupied for prolonged periods of time. Room 6 temperature will be investigated, and probably dropped in favor of Outside Temperature in the Pre-Processing notebook.

````python
hourly_data = data[['TempOutSide', 'TotalConsmp']].resample('H').agg({'TempOutSide': 'mean', 'TotalConsmp': 'sum'})
````

````python
ax = hourly_data.plot(y="TempOutSide", legend=False, figsize=(18, 8))
ax2 = ax.twinx()
hourly_data.plot(y="TotalConsmp", ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
````
